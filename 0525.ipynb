```python
import datetime

start_date = datetime.datetime(2023, 2, 21, 12, 0, 0)
end_date = datetime.datetime(2023, 5, 2, 12, 0, 0)

date_range = pd.date_range(start=start_date, end=end_date, freq='D')

for date in date_range:
    start_date_str = date.strftime('%Y-%m-%d 12:00:00')
    end_date_str = (date + datetime.timedelta(days=1)).strftime('%Y-%m-%d 12:00:00')

    filtered_data = data[(data['Start'] >= start_date_str) & (data['Start'] <= end_date_str)]
    filtered_data = filtered_data[((filtered_data['Source'] == 'NHC') & (filtered_data['Value'] == 'InBed'))] #Inbed 이면서 NHC인것들 

    filtered_data["year-month-day"] = date.strftime('%Y-%m-%d')
    filtered_data["weekday"] = date.weekday()

    # 현재 날짜에 해당하는 필터링된 데이터 저장
    #file_name = f"data_{date.strftime('%Y%m%d')}.csv"
    #file_name = f"data_{start_date_str}.csv"
    file_name = f"data_{date.strftime('%Y_%m_%d')}.csv"
    filtered_data.to_csv(file_name, index=False)
```
